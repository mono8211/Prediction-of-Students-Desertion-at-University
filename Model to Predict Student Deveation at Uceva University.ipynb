# Cleaning Dataframe

```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn as skl
sns.set_theme(style="ticks")
import warnings
plt.rcParams['figure.figsize']= 8,4
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
#loading dataset
df = pd.read_excel('df.xlsx')
```

```python
#first sight of the data
df.info()
```

```python
#Reviewing the dataframe
df.head()
```

```python
#Identifing column names, looking for typos
df.columns
```

```python
#Changing column names
df.rename(columns={'Genero(F=1, M=0)':'Genero','Ciudad de Residencia':'Ciudad_Residencia',\
                   'Motivo Retiro':'Motivo_Retiro','Estado (Desertor=1, No Desertor=0)':'Estado',\
                  'Trabaja actualmente (Si=1, No=0)':'Trabaja_actualmente',\
                  'Graduado industrial (Si=1, No=0)':'Graduado','Resultado Pruebas Admision':'Resultado_Pruebas_Admision', \
                  'Materias Perdidas':'Materias_Perdidas', 'Numero de Apoyos':'Numero_Apoyos', \
                  'Apoyos Academicos':'Apoyos_Academicos', 'Apoyos Financieros':'Apoyos_Financieros', \
                  "Número personas a cargo":'Numero_personas_cargo', 'Periodo ':'Periodo', \
                  'Semestre ':'Semestre'},inplace=True)

df.head()
```

```python
#deleting columns
df.drop(['Motivo_Retiro','Graduado', 'Plan','Periodo'], axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
#Setting independent variable at last position
Desertion_State = df[['Estado']]
df.drop(['Estado'], axis=1, inplace=True)
df.info()
```

```python
df.insert(12,'Estado',Desertion_State)
```

```python
# Setting Variables
df.info()
len(df.Ciudad_Residencia)
df['Ciudad_Residencia'] = df['Ciudad_Residencia'].str.lower()
df['Ciudad_Residencia'] = df['Ciudad_Residencia'].str.replace(r'\s+', '_', regex=True)
df.Ciudad_Residencia.unique()

    
```

```python
#changing values name to eliminate characters

df.loc[df['Ciudad_Residencia']!= 'tuluá','Ciudad_Residencia']='Otra'
df.loc[df['Ciudad_Residencia']== 'tuluá','Ciudad_Residencia']='Tulua'
```

```python
df['Salario'] = df['Salario'].str.lower()
df['Salario'] = df['Salario'].str.replace(r'\s+', '_', regex=True)
```

```python
df.Salario.unique()
```

```python
df.loc[df['Salario']== 'mas_de_8_salarios_minimos','Salario']='Mas de 4'
df.loc[df['Salario']== 'mas_de_4_y_menos_de_6_salarios_minimos','Salario']='Mas de 4'
df.loc[df['Salario']== 'mas_de_6_y_menos_de_8_salarios_minimos','Salario']='Mas de 4'
df.loc[df['Salario']== 'mas_de_2_y_menos_de_4_salarios_minimos','Salario']='Menos de 4'
df.loc[df['Salario']== 'dos_salarios_mínimos_o_menos','Salario']='Menos de 4'
df.loc[df['Salario']== 'no_definido','Salario']='Menos de 4'
```

```python
#encoding the dependent variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

```python
#Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
